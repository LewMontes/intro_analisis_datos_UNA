```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv("casasBoston.csv")
```

```python
print(datos)
```

```python
df = datos[["RM","CRIM","MEDV","TOWN","CHAS"]]
```

```python
df=datos.rename(columns={
    "TOWN":"CIUDAD",
    "CRIM":"INDICE_CRIMEN",
    "INDUS":"PCT_ZONA_INDUSTRIAL",
    "CHAS": "RIO CHARLES",
    "RM": "N_HABITACIONES_MEDIO",
    "MEDV": "VALOR_MEDIANO",
    "LSTAT": "PCT_CLASE_BAJA"
})
```

```python
print(df)
```

```python
df.N_HABITACIONES_MEDIO.plot.hist()
plt.title("Histograma de frecuencia de número de habitaciones")
plt.ylabel("Frecuencia")
plt.xlabel("Número de habitaciones")
plt.show()
```

```python
df.plot.scatter(x="VALOR_MEDIANO", y="INDICE_CRIMEN",alpha=1)
plt.title("Gráfico de dispersión del indice de crimen")
plt.xlabel("Promedio")
plt.ylabel("Índice de crimen")
plt.show()
```

```python
valor_por_ciudad=df.groupby("CIUDAD")["VALOR_MEDIANO"].mean()
```

```python
valor_por_ciudad.head(10).plot.barh()
plt.title("Gráfico de barras de las primeras 10 ciudades por índice de crímen")
plt.xlabel("Promedio del índice de crimen")
plt.ylabel("Ciudad")
plt.show()
```
