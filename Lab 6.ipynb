#Problema

##Se presenta el problema de la reserva del juego "Elden Ring: Shadow of The Erdtree" donde se registran los usuarios y se les da prioridad por la fecha de registro, este registro de hace por medio de CSV

##La soluci√≥n se presenta en Python, tomando el CSV y ordenando la lista por la fecha de registro de cada cliente

```python
import datetime
import csv
from PIL import Image

img = Image.open('image.jpg')
display(img)

waiting_list = []
with open('lista_espera.csv', 'r') as file:
    reader = csv.reader(file)
    next(reader)  # Saltar la cabecera
    for row in reader:
        waiting_list.append({'name': row[0], 'date': datetime.datetime.strptime(row[1], '%Y-%m-%d')})

waiting_list.sort(key=lambda x: x['date'])

now = datetime.datetime.now()
target_date = datetime.datetime(now.year, 6, 21)

if now > target_date:
    target_date = target_date.replace(year=now.year + 1)

days_until_target = (target_date - now).days

print(f"\nQuedan {days_until_target} d√≠as hasta el 21 de Junio, {target_date.year}, para el lanzamiento de Elden Ring: Shadow of The Erdtree\n")

print(f"Actualmente hay {len(waiting_list)} personas en la lista de espera.")

#print("Lista de espera:")
for i, person in enumerate(waiting_list):
    print(f"{i+1}. {person['name']} - {person['date'].strftime('%Y-%m-%d')}")

```
